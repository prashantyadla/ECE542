```python
textobj = []
for d in datdata:
    oldimg = d[:-4].split("_")[-1]
    img = data[oldimg]
    h,w,c = cv2.imread("Data//"+str(img)).shape
    label = oldimg[0].split('\"')[-1][:-1]
    with open(d, 'r') as f:
        temp=f.readlines()
        lol = temp[0][2:-3].replace('""','"').replace('},{','},,{').split(",,")
        for jobj in lol:
            try:
                vals = json.loads(jobj)
            except:
                break
            key = "collar" if "collar" in vals.keys() else "leaf"
            arr = vals[key]
            coordinates = str(round(arr[0][0]))+","+str(round(arr[0][1]))+","+str(round(arr[-1][0]))+","+str(round(arr[-1][1]))
            toadd = str(img)+ "," + str(w) + "," + str(h)+ "," + key + "," + coordinates
            toadd = toadd.replace("[","").replace("]","")
            textobj.append(toadd)
```

```python
len(textobj)
```

```python
q = open("tfile.txt", "a+")
q.write("filename,width,height,class,xmin,ymin,xmax,ymax\n")
for row in textobj:
    q.write(row)
    q.write("\n")
```
